### step-6 : Droping Unnecessary columns

```python
df.drop('Not_Useful_Column', axis=1, inplace=True)
df.drop('Unnamed: 8', axis=1, inplace=True)
df.drop('Unnamed: 9', axis=1, inplace=True)
```

```python
df
```

### Step-7 : Cleaning Last_Name columns

```python
df['Last_Name']=df['Last_Name'].str.strip("123./")
df
```

### Step-8 : Maintain Data Uniformity

```python
df['Paying Customer'] = df['Paying Customer'].replace({'Yes': 'Y', 'No': 'N'})
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace(r'[^0-9]', '', regex=True)  # Remove non-digit characters
df['Phone_Number'] = df['Phone_Number'].str.slice(0, 3) + '-' + df['Phone_Number'].str.slice(3, 6) + '-' + df['Phone_Number'].str.slice(6, 10)
df['Phone_Number']=df['Phone_Number'].fillna(' ')
df
```

```python
df.replace('N/a',np.nan,inplace=True)
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace({'Yes': 'Y', 'No': 'N'})
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].fillna('Y')
df
```

```python
for x in df.index:
    if df.loc[x,"Do_Not_Contact"]=='Y':
        df.drop(x,inplace=True)
df
```

```python
df=df[df['Phone_Number']!='']
df
```

```python
df = df[df['Phone_Number'] != " "]
df
```

```python
df.reset_index(drop=True, inplace=True)
df
```
